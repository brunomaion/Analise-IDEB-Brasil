```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import sys
import joblib

tipodoaquivo = joblib.load('tipodoarquivo.pkl')
display(tipodoaquivo)
```

```python
dfMeso = pd.read_csv("planilhas/MesoPR.csv", sep=";")
total_linhas = len(dfMeso)
dfMeso.rename(columns={'municipio': 'Região'}, inplace=True)
display(dfMeso)

dfBrasilOeste = pd.read_csv("planilhas/BrasilPrOeste.csv", sep=";")
total_linhas = len(dfBrasilOeste)
display(dfBrasilOeste)
```

# ARPOVAÇÃO TOTAL

```python

def graficoLinhaAprovTotal(df, titulo, nomeSalvarArquivo):
    # Configurar o gráfico
    plt.figure(figsize=(12, 6))

    # Plotar cada Região
    for index, row in df.iterrows():
        plt.plot(['2017', '2019', '2021', '2023'], 
                [row['TXApro_Total_2017'], row['TXApro_Total_2019'], row['TXApro_Total_2021'], row['TXApro_Total_2023']],
                label=row['Região'])

    # Adicionar títulos e labels
    plt.title(titulo)
    plt.xlabel('Ano')
    plt.ylabel('Taxa de Aprovação Total')
    plt.legend(title='Região', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.yticks(range(70, 101, 5))

    # Ajustar o layout
    plt.tight_layout()

    nome = 'graficos/meso/' + 'linha'+ 'AprovTotal' +nomeSalvarArquivo + tipodoaquivo
    plt.savefig(nome, dpi=300, bbox_inches = "tight")
    # Mostrar o gráfico
    plt.show()
```

```python
graficoLinhaAprovTotal(dfMeso, 'Mesorregiões do Paraná', 'mesoPR')
graficoLinhaAprovTotal(dfBrasilOeste, 'Região Oeste', 'BraPrOeste')

```
