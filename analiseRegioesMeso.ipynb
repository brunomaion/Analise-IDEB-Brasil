```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import sys
import joblib

tipodoaquivo = joblib.load('tipodoarquivo.pkl')
display(tipodoaquivo)
```

```python
dfMeso = pd.read_csv("planilhas/MesoPR.csv", sep=";")
total_linhas = len(dfMeso)
dfMeso.rename(columns={'municipio': 'Regi√£o'}, inplace=True)
display(dfMeso)

dfBrasilOeste = pd.read_csv("planilhas/BrasilPrOeste.csv", sep=";")
total_linhas = len(dfBrasilOeste)
display(dfBrasilOeste)
```

# ARPOVA√á√ÉO TOTAL

```python

def graficoLinhaAprovTotal(df, titulo, nomeSalvarArquivo):
    # Configurar o gr√°fico
    plt.figure(figsize=(12, 6))

    # Plotar cada Regi√£o
    for index, row in df.iterrows():
        plt.plot(['2017', '2019', '2021', '2023'], 
                [row['TXApro_Total_2017'], row['TXApro_Total_2019'], row['TXApro_Total_2021'], row['TXApro_Total_2023']],
                label=row['Regi√£o'])

    # Adicionar t√≠tulos e labels
    plt.grid()
    plt.title(titulo)
    plt.xlabel('Ano')
    plt.ylabel('Taxa de Aprova√ß√£o Total')
    plt.legend(title='Regi√£o', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.yticks(range(70, 101, 5))

    # Ajustar o layout
    plt.tight_layout()

    nome = 'graficos/meso/' + 'linha'+ 'AprovTotal' +nomeSalvarArquivo + tipodoaquivo
    plt.savefig(nome, dpi=300, bbox_inches = "tight")
    # Mostrar o gr√°fico
    plt.show()
```

```python
graficoLinhaAprovTotal(dfMeso, 'Mesorregi√µes do Paran√°', 'mesoPR')
graficoLinhaAprovTotal(dfBrasilOeste, 'Regi√£o Oeste', 'BraPrOeste')

```

# BARRA

```python
import matplotlib.pyplot as plt
import numpy as np

def graficoBarra20171(df, titulo, nomeSalvarArquivo, coluna):
  
    # Configurar o gr√°fico
    plt.figure(figsize=(12, 6))

    # Plotar os dados de 2017 para cada Regi√£o
    bars = plt.bar(df['Regi√£o'], coluna, color='skyblue')

    # Adicionar t√≠tulos e labels
    plt.grid(axis='y')
    plt.title(titulo)
    plt.xlabel('Regi√£o')
    plt.ylabel(str(coluna.name))
    plt.xticks(rotation=50)

    # Adicionar os valores no topo das barras
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

    # Ajustar o layout
    plt.tight_layout()

    # Salvar o gr√°fico
    nome = 'graficos/meso/' + 'barra' + str(coluna.name) + nomeSalvarArquivo + '.png'
    plt.savefig(nome, dpi=300, bbox_inches="tight")

    # Mostrar o gr√°fico
    plt.show()

def graficoBarra20172(df, titulo, nomeSalvarArquivo, coluna):
  
    # Configurar o gr√°fico
    plt.figure(figsize=(12, 6))

    # Plotar os dados de 2017 para cada Regi√£o
    bars = plt.bar(df['Regi√£o'], coluna, color='skyblue')

    # Adicionar t√≠tulos e labels
    plt.grid(axis='y')
    plt.title(titulo)
    plt.xlabel('Regi√£o')
    plt.ylabel(str(coluna.name))
    #plt.xticks(rotation=50)

    # Adicionar os valores no topo das barras
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

    # Ajustar o layout
    plt.tight_layout()

    nome = 'graficos/meso/' + 'barra' + 'barra' + str(coluna.name) + nomeSalvarArquivo + '.png'
    plt.savefig(nome, dpi=300, bbox_inches="tight")

    # Mostrar o gr√°fico
    plt.show()
    

```
