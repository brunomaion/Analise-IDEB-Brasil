# ARPOVAÇÃO TOTAL

```python

def graficoLinhaAprovTotal(df, titulo, nomeSalvarArquivo):
    # Configurar o gráfico
    plt.figure(figsize=(12, 6))

    # Plotar cada Região
    for index, row in df.iterrows():
        plt.plot(['2017', '2019', '2021', '2023'], 
                [row['TXApro_Total_2017'], row['TXApro_Total_2019'], row['TXApro_Total_2021'], row['TXApro_Total_2023']],
                label=row['Região'])

    # Adicionar títulos e labels
    plt.title(titulo)
    plt.xlabel('Ano')
    plt.ylabel('Taxa de Aprovação Total')
    plt.legend(title='Região', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.yticks(range(0, 200, 5))

    # Ajustar o layout
    plt.tight_layout()

    nome = 'graficos/mesoEP/' + 'linha'+ 'AprovTotal' +nomeSalvarArquivo + tipodoaquivo
    plt.savefig(nome, dpi=300, bbox_inches = "tight")
    # Mostrar o gráfico
    plt.show()
```

```python

graficoLinhaAprovTotal(dfBrasilOeste, 'Região Oeste', 'BraPrOeste')

```

```python
import matplotlib.pyplot as plt
import numpy as np

def graficoBarra20171(df, titulo, nomeSalvarArquivo, coluna):
  
    # Configurar o gráfico
    plt.figure(figsize=(12, 6))

    # Plotar os dados de 2017 para cada Região
    bars = plt.bar(df['Região'], coluna, color='skyblue')

    # Adicionar títulos e labels
    plt.grid(axis='y')
    plt.title(titulo)
    plt.xlabel('Região')
    plt.ylabel(str(coluna.name))
    plt.xticks(rotation=50)

    # Adicionar os valores no topo das barras
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

    # Ajustar o layout
    plt.tight_layout()

    # Salvar o gráfico
    nome = 'graficos/mesoEP/' + 'barra' + str(coluna.name) + nomeSalvarArquivo + tipodoaquivo
    plt.savefig(nome, dpi=300, bbox_inches="tight")

    # Mostrar o gráfico
    plt.show()

def graficoBarra20172(df, titulo, nomeSalvarArquivo, coluna):
  
    # Configurar o gráfico
    plt.figure(figsize=(12, 6))

    # Plotar os dados de 2017 para cada Região
    bars = plt.bar(df['Região'], coluna, color='skyblue')

    # Adicionar títulos e labels
    plt.grid(axis='y')
    plt.title(titulo)
    plt.xlabel('Região')
    plt.ylabel(str(coluna.name))
    plt.xticks(rotation=50)

    # Adicionar os valores no topo das barras
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

    # Ajustar o layout
    plt.tight_layout()

    nome = 'graficos/mesoEP/' +'barra' + str(coluna.name) + nomeSalvarArquivo + tipodoaquivo
    plt.savefig(nome, dpi=300, bbox_inches="tight")

    # Mostrar o gráfico
    plt.show()
    

```
